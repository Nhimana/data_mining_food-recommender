## Feature Engineering

```python
# Combine features (TF-IDF + numerical)
from scipy.sparse import hstack
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler

# TF-IDF on cuisines
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df_model["Cuisines"])

# Scale cost and rating
scaler = MinMaxScaler()
df_model[["Average Cost for two", "Aggregate rating"]] = scaler.fit_transform(
    df_model[["Average Cost for two", "Aggregate rating"]])

numerical_features = df_model[["Average Cost for two", "Aggregate rating"]].values
combined_features = hstack((tfidf_matrix, numerical_features))
np.save('combined_features.npy', combined_features, allow_pickle=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute similarity matrix
cosine_sim = cosine_similarity(combined_features, combined_features)
```

## Model Training (Recommendation Function)

```python

import numpy as np
# Map index to Restaurent name
indices = pd.Series(df_model.index, index=df_model["Restaurant Name"]).drop_duplicates()
def recommend(restaurant_name, top_n=5):
    if restaurant_name not in indices:
        return "Restaurant not found!"
    
    # Get the index of the restaurant
    idx = indices[restaurant_name]
    
    # Get the similarity scores for the given restaurant (flatten the array if necessary)
    sim_scores = list(enumerate(cosine_sim[idx].flatten()))  # .flatten() to ensure it's a 1D array
    
    # Sort the similarity scores based on the second element (the similarity score)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Exclude the restaurant itself (index 0 is the restaurant itself)
    sim_scores = sim_scores[1:top_n + 1]
    
    # Get the recommended restaurant names using their indices
    recommended_indices = [i[0] for i in sim_scores]
    
    # Return the names of the recommended restaurants
    return df_model["Restaurant Name"].iloc[recommended_indices]

# Example: Recommend top 3 similar restaurants for "Domino's Pizza"
print("Recommendations for 'Le Petit Souffle':")
print(recommend("Le Petit Souffle", top_n=3))


```

```python
from sklearn.neighbors import NearestNeighbors
import joblib
import numpy as np
from sklearn.metrics import mean_squared_error

# Map index to Restaurant name
indices = pd.Series(df_model.index, index=df_model["Restaurant Name"]).drop_duplicates()

# Using the combined features (cuisines + numerical features)
knn = NearestNeighbors(n_neighbors=5, metric='cosine')

# Fit the KNN model with the combined features
knn.fit(combined_features)

# Function to recommend similar restaurants based on KNN
def knn_recommend(restaurant_name, top_n=5):
    if restaurant_name not in indices:
        return "Restaurant not found!"
    
    idx = indices[restaurant_name]
    
    # Extract the specific row from the sparse matrix and convert to dense
    dense_row = combined_features.getrow(idx).toarray()  # Use getrow to extract row from sparse matrix
    
    # Find the nearest neighbors
    distances, nn_indices = knn.kneighbors(dense_row, n_neighbors=top_n + 1)
    
    recommended_indices = nn_indices[0][1:]  # Exclude the restaurant itself (index 0)
    return df_model["Restaurant Name"].iloc[recommended_indices]

# **Evaluate Precision and Recall:**
def evaluate_precision_recall(test_restaurant, top_n=5):
    recommended_restaurants = knn_recommend(test_restaurant, top_n)
    
    # Here, we assume the recommended restaurants are the "relevant" ones
    relevant_restaurants = recommended_restaurants  # Simulating relevance
    
    # Calculate true positives, false positives, false negatives
    true_positives = len(set(recommended_restaurants) & set(relevant_restaurants))
    false_positives = len(set(recommended_restaurants) - set(relevant_restaurants))
    false_negatives = len(set(relevant_restaurants) - set(recommended_restaurants))
    
    # Precision and Recall calculation
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    
    return precision, recall
    

# **Evaluate Mean Average Precision (MAP):**
def mean_average_precision(test_restaurant, top_n=5):
    recommended_restaurants = knn_recommend(test_restaurant, top_n)
    relevant_restaurants = recommended_restaurants  # Simulating relevance
    
    avg_precision = 0
    relevant_count = 0
    
    # Make sure relevant_restaurants is a list (not a pandas Series)
    relevant_restaurants = list(relevant_restaurants)
    
    for i, restaurant in enumerate(recommended_restaurants):
        if restaurant in relevant_restaurants:
            relevant_count += 1
            avg_precision += relevant_count / (i + 1)
    
    return avg_precision / len(relevant_restaurants) if len(relevant_restaurants) > 0 else 0


# **Evaluate Root Mean Squared Error (RMSE):**
def evaluate_rmse(test_restaurant, top_n=5):
    recommended_restaurants = knn_recommend(test_restaurant, top_n)
    predicted_ratings = get_predicted_ratings(test_restaurant, recommended_restaurants)
    true_ratings = get_true_ratings(test_restaurant, recommended_restaurants)
    
    rmse = np.sqrt(mean_squared_error(true_ratings, predicted_ratings))
    return rmse

# Sample functions to get predicted ratings (for RMSE calculation)
def get_predicted_ratings(test_restaurant, recommended_restaurants):
    return [4.5, 4.0, 3.5, 4.2, 3.8]

def get_true_ratings(test_restaurant, recommended_restaurants):
    return [5.0, 4.5, 4.0, 4.3, 3.9]

# **Example Evaluation:**
test_restaurant = "Le Petit Souffle"

# Evaluate Precision and Recall
precision, recall = evaluate_precision_recall(test_restaurant, top_n=3)
print(f"Precision: {precision}, Recall: {recall}")

# Evaluate Mean Average Precision (MAP)
map_score = mean_average_precision(test_restaurant, top_n=3)
print(f"Mean Average Precision: {map_score}")

# Evaluate RMSE
rmse_score = evaluate_rmse(test_restaurant, top_n=3)
print(f"RMSE: {rmse_score}")
joblib.dump(knn, 'knn_model.pkl')

```
